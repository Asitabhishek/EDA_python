**how to deal with outliers**
- 1. we can drop the outliers
  - Droping outlier is an easy task,but not recommended
  - if youndrop the outliers we are dropping the row
  - which means we are droping the other columns value also
  - may be some important we might loss from other column values
  - but in industry if outlier percentage less than 2% of the total data ,we can drop
  - the number of outlier in pwage is 427
  - the number of total observation are 25480
  - the percentage is 1.6 less than 2%
  2. we already know that outliers does not affect by median
  - so it is the best pratice to fill the outliers with median
  3. capping or flooring
  - insted of fill with median
  - we can fill with certain values
  - the outliers which are less than LB fill with LB value
  - the outlier which are greater than Ub fill with LB
- isolation Forest
  - is ML model is used to find the outlier and anamolies  


**Fill with Median Value**

**np.where**
- in numpy we have method called where
- it takes three arguments
  1. Condition
  2. value if condition is True
  3. value if condition is False
- it is applicable for only binary means if and else

```python
list1=[]
for value in wage_data:
    if value < LB or value > UB:
        median= wage_data.median()
        list1.append(median)
    else:
        list1.append(value)
```
